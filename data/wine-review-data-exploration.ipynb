## Part 1 - Focus on General Data Distribtutions

```python
data = pd.read_csv("wine-reviews/winemag-data-130k-v2.csv")

print("Number of entries")
len(data)
```

```python
data.describe().T
```

```python
data.dtypes
```

```python
varieties = data["variety"].value_counts()
countries = data["country"].value_counts()
provinces = data["province"].value_counts()
wineries = data["winery"].value_counts()


for x in [("varities", varieties), ("countries", countries), ("provinces", provinces)]:
    print("######################################")
    print("Number of", x[0], len(x[1]))
    print("Entry stats - max:", x[1].max(), "| min:", x[1].min(), "| median:", x[1].median())
    print("Top", x[0])
    print(x[1].head())
    print()
```

```python
varieties.hist()
```

```python
varieties.head(20).plot(kind ="bar", title = "Number of entries for top 20 varieties")
```

```python
countries.hist()
```

```python
countries.head(20).plot(kind ="bar",title = "Number of entries for top 20 countries")
```

```python
provinces.hist()
```

```python
provinces.head(20).plot(kind ="bar",title = "Number of entries for top 20 provinces")
```
